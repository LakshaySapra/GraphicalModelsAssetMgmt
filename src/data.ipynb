```python
ret_periods = [1,5,10,21,42,63,126]
min_periods = [1,5,10,20,40,60,120]
hist_ret_lst = []

for period,min_period in zip(ret_periods,min_periods):
    ser = ret_df.rolling(period,min_periods=min_period,axis=0).mean().stack()*252/period
    ser = ser.rename(f'annret_{period}')
    hist_ret_lst.append(ser)

    ser = ret_df.rolling(period,min_periods=min_period,axis=0).std().stack()*np.sqrt(252/period)
    ser = ser.rename(f'annvol_{period}')
    hist_ret_lst.append(ser)

hist_ret_df = pd.concat(hist_ret_lst,axis=1).drop('annvol_1',axis=1).dropna(axis=0,how='any').sort_index()
hist_ret_df
```

```python
df3 = pd.read_csv('data/supply_chain.csv')
print(df3.columns)
df3
```

```python
supchain_df = df3[['srcdate','gvkey','cgvkey','salecs']].sort_values(['srcdate'])
supchain_df['srcdate'] = pd.to_datetime(supchain_df['srcdate'])
supchain_df.to_pickle('data/supchain_df.pkl',protocol=4)
supchain_df
```

```python
def get_gvkey_dict():
    gvkey_dict = set(supchain_df['gvkey']).intersection(supchain_df['cgvkey'])
    N = len(gvkey_dict)
    gvkey_dict = dict(zip(list(gvkey_dict),range(N)))
    return gvkey_dict

def get_edges(date,gvkey_dict):
    cur_df = supchain_df[(supchain_df['srcdate'] <= date) & (supchain_df['srcdate'] >= date -pd.DateOffset(years=3)) ]
    gvkeys = gvkey_dict.keys()
    cur_df = cur_df[cur_df['gvkey'].isin(gvkeys) & cur_df['cgvkey'].isin(gvkeys)]
    c_edge_list = cur_df[['gvkey','cgvkey']].rename({'gvkey':'sup','cgvkey':'con'},axis=1).applymap(lambda x:gvkey_dict[x])
    s_edge_list = c_edge_list[['con','sup']].rename({'sup':'con','con':'sup'},axis=1)
    
    return c_edge_list.values,s_edge_list.values
```
